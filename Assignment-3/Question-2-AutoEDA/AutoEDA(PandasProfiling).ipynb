# ***Performing Auto Exploratory Data Analysis on CarPrice_Assignment dataset using Pandas Profiler***



 **Using Pandas Profiler below information can be known:**
* Correlations: high correlation warnings, based on different correlation metrics (Spearman, Pearson, Kendall, Cramér’s V, Phik)

* Missing values: through counts, matrix, heatmap and dendrograms

* Duplicate rows: list of the most common duplicated rows

* Text analysis: most common categories (uppercase, lowercase, separator), scripts (Latin, Cyrillic) and blocks (ASCII, Cyrilic)

```python
import pandas as pd
import numpy as np
from pandas_profiling import ProfileReport
```

```python
df = pd.read_csv('/content/CarPrice_Assignment.csv')
```

```python
df.shape
```

*  ***For better readability replace spaces in all the columns with '_' and also convert in to lower case***

```python
strings = list(df.dtypes[df.dtypes == 'object'].index)
for s in strings:
  df[s] = df[s].str.lower().str.replace(' ', '_')

df.head()
```

* ***Convert all column titles to lower case and replace whitespace with '_'***

```python
df.columns = df.columns.str.lower().str.replace(' ', '_')
df.columns
```

```python
df.head()
```

```python
pandas_profile = ProfileReport(df, title='Pandas Profiling report for Car price dataset', explorative = True)
```

```python
pandas_profile.to_file("output_car_price_pandasProfiling.html")
```
