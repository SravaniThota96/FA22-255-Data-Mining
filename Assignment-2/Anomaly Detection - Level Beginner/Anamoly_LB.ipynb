```python
import pandas as pd

dataset = pd.read_csv('wine-clustering.csv')
```

```python
data = dataset.sample(frac=0.95, random_state=786)
data_unseen = dataset.drop(data.index)

data.reset_index(drop=True, inplace=True)
data_unseen.reset_index(drop=True, inplace=True)

print('Data for Modeling: ' + str(data.shape))
print('Unseen Data For Predictions: ' + str(data_unseen.shape))
```

```python
from pycaret.anomaly import *

anom = setup(data, normalize = True, 
                   session_id = 255)
```

```python
models()
```

```python
abod = create_model('abod')
```

```python
svm = create_model('svm')
```

```python
svm_results = assign_model(svm)
```

```python
svm_results.head()
```

```python
abod_results = assign_model(abod)
```

```python
plot_model(abod)
```

```python
plot_model(abod, plot='umap')
```

```python
data_predictions = predict_model(abod, data = data)
data_predictions.head()
```
